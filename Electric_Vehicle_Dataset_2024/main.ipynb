```python
# Remove Irrelevant Columns
df = df.drop(columns={"Base MSRP", "Legislative District", "DOL Vehicle ID", "Vehicle Location", "VIN (1-10)",})
```

```python
# Pie Chart: BEV and PHEV

df['Electric Vehicle Type'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.show()
```

```python
# Count Carbrand and store into dataframe
carbrand = df['Make'].value_counts().reset_index()
carbrand.style.background_gradient(cmap="CMRmap_r")
```

```python
df["Clean Alternative Fuel Vehicle (CAFV) Eligibility"].value_counts()
```

```python
# Distribution of E Vehicles by County

# Group County's and count e vehicles in each county
county_count = df.groupby('County').size().reset_index(name='Electric Vehicle Count')
# county_count.head()
# Sort counties by electric vehicle count in descending order
county_count = county_count.sort_values(by='Electric Vehicle Count', ascending=False)
county_count=county_count.head(10)
# Plot distribution of e vehicles by county
plt.figure(figsize=(10, 6))
plt.bar(county_count['County'], county_count['Electric Vehicle Count'], color='blue')
plt.xlabel('County')
plt.ylabel('Electric Vehicle Count')
plt.title('Distribution of Electric Vehicle by County')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
df['Electric Vehicle Type'].value_counts().plot(kind='bar')
plt.title('Electric Vehicle Type Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# E-Vehicle by year count
EVehicle_by_year = df.groupby('Model Year').size().reset_index(name='Electric Vehicle Count')

# Plot E Vehicle Manufacturing overtime
plt.figure(figsize=(10,6))
plt.plot(EVehicle_by_year["Model Year"],EVehicle_by_year["Electric Vehicle Count"], marker='o', color='blue')
plt.xlabel('Model Year')
plt.ylabel("Electric Vehicle Count")
plt.title('Distribution of Electric Vehicles by Year Over Time')
plt.grid(True)
plt.show()
```

```python
# Define Eligible Vehicles
eligible_vehicles = df[df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'] == 'Yes']

# Group by Eligibility status and count number of vehicles
eligibility_count = df.groupby("Clean Alternative Fuel Vehicle (CAFV) Eligibility").size()

# Plot the comparison of adoption rates between eligible and non-eligible vehicles
plt.figure(figsize=(8,6))
eligibility_count.plot(kind="pie", autopct='%1.1f%%', colors=['lime','yellow', 'red'])
plt.title('Adoption Rates of Eligible vs. Non-Eligible Vehicles')
plt.show()
```

```python
# Histogram for Distribution of electric range values

plt.figure(figsize=(10,6))
sns.histplot(df['Electric Range'], bins=20, kde=True, color='lime')
plt.xlabel('Electric Range (Miles)')
plt.ylabel('Frequency')
plt.title('Distribution of Electric Range for E-Vehicles')
plt.show()
```

```python
# Plot a bar chart to visualize the distribution of electric vehicles by utility providers
electric_utility_count = df['Electric Utility'].value_counts().head(8)
plt.figure(figsize=(10,6))
electric_utility_count.plot(kind='bar', color='blue')
plt.xlabel('Electric Utility Provider')
plt.ylabel('Number of Electric Vehicles')
plt.title('Distribution of E Vehicle Adoption by Utility Provider')
plt.xticks(rotation=45, ha='right')
plt.show()
```
